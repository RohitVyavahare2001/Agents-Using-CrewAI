```python
from crewai import Agent,LLM,Crew,Process
from crewai_tools import SerperDevTool
from crewai import Task
```

```python
import os 
from dotenv import load_dotenv
load_dotenv()
os.environ["groq_api_key"]=os.getenv('groq_api_key')
```

```python
Serper_api_key=os.getenv('serper_api_key')
```

```python
llm=LLM(model="groq/gemma2-9b-it")
```

```python
from crewai_tools import ScrapeWebsiteTool
```

```python
search_tool=ScrapeWebsiteTool()
search_tool2=SerperDevTool()
```

```python
travel_researcher=Agent(
    role="Searches for travel destinations, activities, and accommodations based on user preferences",
    goal="""Given a travel {destination} and the number of {days} the user wants to travel for, first generate a list of 3 search terms related to that destination and the number of days,
            For each search term, `search_google` and analyze the results.
            From the results of all searches, return the 10 most relevant results to the user's preferences.
            Remember: the quality of the results is important.""",
    llm=llm,
    tools=[search_tool,search_tool2],
    verbose=False,
    memory=True,
    backstory="""
        You are a world-class travel researcher. Given a travel destination and the number of days the user wants to travel for,
        generate a list of search terms for finding relevant travel activities and accommodations.
        Then search the web for each term, analyze the results, and return the 10 most relevant results.
        """,
allow_delegation=True
)
```

```python
researcher_task = Task(
    description="Follow up with travel_researcher and help him carry out desired tasks",
    expected_output="return the 10 most relevant results to the user's preferences",
    agent=travel_researcher,
    tools=[search_tool, search_tool2]
)
```

```python
planner=Agent(
    role="Generates a draft itinerary based on user preferences and research results",
    goal="""You are a senior travel planner. Given a travel destination, the number of days the user wants to travel for, and a list of research results,
        your goal is to generate a draft itinerary that meets the user's needs and preferences.
        """,
    llm=llm,
    backstory="""
    Given a travel destination, the number of days the user wants to travel for, and a list of research results, generate a draft itinerary that includes suggested activities and accommodations.
    Ensure the itinerary is well-structured, informative, and engaging.
    Provide a nuanced and balanced itinerary, quoting facts where possible.
    Remember: the quality of the itinerary is important.
    Focus on clarity, coherence, and overall quality.
    Never make up facts or plagiarize. Always provide proper attribution.
""",
    verbose=False,
    memory=True
)
```

```python
# Task for Planner Agent
planner_task= Task(
description="follow up with the planner and help him out",
expected_output="generate a draft itinerary that meets the user's needs and preferences.",
agent=planner,
)
```

```python
crew_instance=Crew(
    agents=[travel_researcher,planner],
    tasks=[researcher_task,planner_task],
    process=Process.sequential
)
```

```python
result=crew_instance.kickoff(inputs={"destination":"manali","days":"10"})
print(result)
```
